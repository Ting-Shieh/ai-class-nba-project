```python
url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'.format(year)
   
r = requests.get(url)
r_html = r.text
soup = BeautifulSoup(r_html,'html.parser')
```

```python
table=soup.find_all(class_="full_table")
    
""" Extracting List of column names"""
head=soup.find(class_="thead")
column_names_raw=[head.text for item in head][0]
column_names_polished=column_names_raw.replace("\n",",").split(",")[2:-1]
```

```python
print(column_names_polished)
```

```python
players=[]
    
for i in range(len(table)):

    player_=[]

    for td in table[i].find_all("td"):
        player_.append(td.text)
    
    players.append(player_)
df=pd.DataFrame(players, columns=column_names_polished).set_index("Player")
#cleaning the player's name from occasional special characters
df.index=df.index.str.replace('*', '')
```

#### [從資料分析到資料產品]：DAY 3 資料產品實做-Content Based Recommendation Flask API-相似NBA球員推薦

```python
#抓取球員資料函數
def get_stats(year):
    url='https://www.basketball-reference.com/leagues/NBA_' + year +'_per_game.html'
    nba=pd.read_html(url)[0]
    nba['year']=year
    return nba

# 取得2018年球員資料
nba=get_stats('2018')
```

```python
nba
```

```python
nba.columns
```

```python
nba.groupby(['Player', 'year'])['Rk'].transform('count')
```

```python
#刪除當年度季中轉隊球員重複資料
nba['filter_1'] = nba.groupby(['Player', 'year'])['Rk'].transform('count')
nba.head()
```

```python
# #移除nba中重複的欄位名稱，如：Rk、AST、TM
nba=nba[((nba.Age != 'Age') & ((nba.filter_1==1) | (nba.Tm=='TOT')))]
```

```python
nba.head(25)
```

```python
"""
Index(['Rk', 'Player', 'Pos', 'Age', 'Tm', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'year'],
      dtype='object')
"""
# 轉換str為float
for idx in ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'year']:
    nba[idx]=nba[idx].astype('float')
#     nba[idx]=pd.to_numeric(nba[idx])

# 篩除出賽數不足15場的球員
nba=nba[(nba.G>=15) ]
# 重設index
nba=nba.reset_index(drop=True)
```

```python
nba
```

```python
#比率項目加權
for idx in ['FG', 'FGA', '3P', '3PA','FT', 'FTA']:
    nba[idx+'_m'] = nba.groupby(['year'])[idx].transform('mean')
nba['FG%_m']=nba['FG_m']/nba['FGA_m']
nba['3P%_m']=nba['3P_m']/nba['3PA_m']
nba['FT%_m']=nba['FT_m']/nba['FTA_m']
nba['FG%_w']=(nba['FG%']-nba['FG%_m'])*nba['FGA']
nba['3P%_w']=(nba['3P%']-nba['3P%_m'])*nba['3PA']
nba['FT%_w']=(nba['FT%']-nba['FT%_m'])*nba['FTA']
```

```python
nba
```

```python
# 數據標準化(Z-score)
def z_score(data): 
    nba[data+'_m']=nba.groupby(['year'])[data].transform('mean')
    nba[data+'_s']=nba.groupby(['year'])[data].transform('std')
    nba[data+'_z']=(nba[data]-nba[data+'_m'])/nba[data+'_s']
    del (nba[data+'_m'], nba[data+'_s'])
for idx in ['FG%_w', '3P%_w', 'FT%_w', 'TRB', 'AST', 'BLK', 'STL', 'PTS', 'PF', 'TOV']:
    z_score(idx)
```

```python
nba
```

```python
nba_similar =nba.copy()
```

```python
import numpy as np
# 定義餘弦相似性函數
def cos_similar(x, y):
    similar_score=x.dot(y)/(np.sqrt(x.dot(x))*np.sqrt(y.dot(y)))
    return similar_score
```

```python
#計算各球員近似球員TOP5
player_list=pd.DataFrame({"Player" :[],
            "Pos" :[],
            "FG%_w_z" :[],
            "3P%_w_z" :[],
            "FT%_w_z" :[],
            "TRB_z" :[],
            "AST_z" :[],
            "BLK_z" :[],
            "STL_z" :[],
            "PTS_z" :[],
            "PF_z" :[],
            "TOV_z" :[],
            "player_compared" :[],
            "similar_score" :[]
        })
for idx_1 in nba_similar.index:
    similar=[]
    play=np.array(nba_similar.loc[idx_1,'FG%_w_z':'TOV_z'])
    for idx_2 in nba_similar.index:
        similar.append(cos_similar(play, np.array(nba_similar.loc[idx_2,'FG%_w_z':'TOV_z'])))
    play=(nba_similar.loc[:,'Player':'TOV_z'])
    play['player_compared']=str((nba_similar.values[idx_1][0]))
    play['similar_score']=similar 
    play=(play.sort_values(by='similar_score', ascending=False)).iloc[1:6,:]
player_list=pd.concat([player_list, play],axis=0)
```
