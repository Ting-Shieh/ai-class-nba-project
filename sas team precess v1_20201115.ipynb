```python
sas_games_df = games_df.loc[(games_df['HOME_TEAM_ID']==sasId)|(games_df['VISITOR_TEAM_ID']==sasId)]

sasYearDF = {}

for year in all_year_list[0:6]:
    
    sasYearDF[str(year)+'df'] = sas_games_df.loc[(sas_games_df['SEASON'].isin(all_year_list[0:6]))&(sas_games_df['SEASON']==year)]
```

```python
# tag sas 贏
def tagSasWin(homeid,awayid,home_win):
    if homeid==sasId and home_win==1:

        return 1
    elif awayid==sasId and home_win==0:
        return 1
    else:
        return 0
```

```python
# 每年的sas比賽場次名單
for itemKey in sasYearDF.keys():
    print(itemKey)
    # tag 此場球賽是否贏
    sasYearDF[itemKey]['isSASWin'] = sasYearDF[itemKey].apply(lambda x : tagSasWin(x['TEAM_ID_home'],x['TEAM_ID_away'],x['HOME_TEAM_WINS']), axis=1)
#     sasYearDF[itemKey].to_csv(f'{itemKey}_game_data.csv')
    gameList = sasYearDF[itemKey].loc[:,'GAME_ID'].unique().tolist()
    # 抓出有比賽球員數據
    gameThreeLayer = games_details_df.loc[(games_details_df['GAME_ID'].isin(gameList))&(games_details_df['TEAM_ID']==sasId)].set_index(['GAME_ID', 'PLAYER_ID', 'START_POSITION'])
#     gameThreeLayer.to_csv(f'{itemKey}_game_details_data.csv')
```

```python
"""
GP, GS: games played, games started (出場數，先發場數)
PTS: points(得分)
+ / - : 在場上時隊得分與失分差距。=>例如，在得分平手時上場，打一陣子下場時，球隊贏10分，
FGM, FGA, FG%: field goals made, attempted and percentage(投中次數，投籃次數，百分比)
FTM, FTA, FT%: free throws made, attempted and percentage(罰球命中次數，罰球次數，百分比)
3FGM, 3FGA, 3FG%: three-point field goals made, attempted and percentage(3分球命中次數，出手次數，百分比)
REB, OREB, DREB: rebounds, offensive rebounds, defensive rebounds(籃板球，進攻，防守籃板球)
AST: assists(助攻)

STL: steals(抄截)

BLK: blocks(火鍋)

BA: (block against) 吃火鍋

BS: (blocked shots) 賞火鍋

TO: turnovers(失誤)
EFF: efficiency: NBA's efficiency rating: (PTS + REB + AST + STL + BLK - FG missed - FT missed - TO)

NBA效率 = (得分+籃板+助攻+抄截+火鍋 - 投籃失手 - 罰球失手 - 失誤)

PF: personal fouls(個人犯規)

MIN: minutes(上場分鐘數)

F - 前鋒
C - 中鋒
G - 後衛
G-F - 後衛型前鋒
F-G - 小前鋒
G - 上場次數
HI - 本賽季單場最高得分
MIN - 總出場時間
MPG - 每場得分
PF - 犯規次數
PPG - 平均得分
PTS - 總得分
ST - 搶斷
"""
games_details_df.columns
```
